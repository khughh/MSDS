# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data was taken from Kaggle Dataset
#https://www.kaggle.com/datasets/fireballbyedimyrnmom/us-counties-covid-19-dataset/data
#https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv

data = pd.read_csv('covid.csv')
data
```

```python
wadata = data[data['state'] == 'Washington']

countydeath = wadata.groupby('county')['deaths'].sum().reset_index()

sort = countydeath.sort_values(by='deaths', ascending=False)
sort
```

```python
wadatasort = wadata.sort_values(by=['county', 'date'])
latest = wadatasort.drop_duplicates(subset=['county'], keep='last')
latest
```

```python
latestsort = latest.sort_values(by='deaths', ascending=False)
```

```python
plt.figure(figsize=(18, 16))
plt.pie(latestsort['deaths'], labels=latestsort['county'], autopct='%1.1f%%', startangle=90)
plt.title('Most Recent COVID-19 Deaths by County in Washington State')
plt.show()
```

```python
latest['death_ratio'] = latest['deaths'] / latest['cases']
latestsort2 = latest.sort_values(by='death_ratio', ascending=False).head(20)
latestsort2
```

```python
plt.figure(figsize=(16,14))
plt.pie(latestsort2['death_ratio'], labels=latestsort2['county'], startangle=90)
plt.title('Deaths to Cases Ratio for Top 20 Counties in Washington State')
plt.show()
```

```python
labels = [f"{row['county']} (Ratio: {row['death_ratio']:.2%})" for index, row in latestsort2.iterrows()]

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(latestsort2['deaths'], labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('COVID-19 Deaths and Death/Case Ratio for Top 20 Counties in Washington State')
plt.show()
```

```python
top_20_counties = latestsort.head(20)

plt.figure(figsize=(20, 20))
wedges, texts, autotexts = plt.pie(top_20_counties['deaths'], labels=top_20_counties['county'], autopct='%1.1f%%', startangle=90, textprops={'fontsize': 14})
for text in texts:
    text.set_rotation(20)

plt.legend(wedges, top_20_counties['county'], title="Counties", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Most Recent COVID-19 Deaths by County in Washington State (Top 20 Counties)')
plt.show()
```
