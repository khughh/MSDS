```python
import matplotlib.pyplot as plt

# Given values
x = 0.05
y = 0.90
z = 0.85

# Calculated values
TP = 45
FP = 142
TN = 808
FN = 5
precision = TP / (TP + FP)
recall = TP / (TP + FN)  # same as sensitivity
specificity = TN / (TN + FP)

# ROC Curve
fpr_values = [0, 1 - specificity, 1]
tpr_values = [0, recall, 1]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(fpr_values, tpr_values, marker='o')
plt.plot([0, 1], [0, 1], linestyle='--')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate (1-Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')

# Precision-Recall Curve
recall_values = [0, recall, 1]
precision_values = [precision, precision, TP / (TP + FN)]

plt.subplot(1, 2, 2)
plt.plot(recall_values, precision_values, marker='o')
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')

plt.tight_layout()
plt.show()
```
